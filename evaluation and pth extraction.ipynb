```python
def load_checkpoints(file_path, encoder, encoder_solver, decoder, decoder_solver, refiner=None, refiner_solver=None,
                     merger=None, merger_solver=None):
    print('[INFO] %s Loading checkpoint from %s ...' % (dt.now(), file_path))
    checkpoint = torch.load(file_path,map_location='cpu')
    # Modify the keys in the checkpoint's state dictionary
    modified_state_dict = {}
    modified_state_dict2 = {}
    for key in checkpoint['encoder_state_dict']:
        new_key = key.replace('module.', '') # Remove the 'module.' prefix
        modified_state_dict[new_key] = checkpoint['encoder_state_dict'][key]
        
    for key in checkpoint['decoder_state_dict']:
        new_key = key.replace('module.', '')  # Remove the 'module.' prefix
        modified_state_dict2[new_key] = checkpoint['decoder_state_dict'][key]

    encoder.load_state_dict(modified_state_dict)
    decoder.load_state_dict(modified_state_dict2)
    # Load refiner and merger state dictionaries if available
    if refiner is not None and 'refiner_state_dict' in checkpoint:
        refiner.load_state_dict(checkpoint['refiner_state_dict'],strict=False)
    if merger is not None and 'merger_state_dict' in checkpoint:
        merger.load_state_dict(checkpoint['merger_state_dict'],strict=False)

    # Return the loaded checkpoint
    return checkpoint['epoch_idx'], checkpoint['best_iou'], checkpoint['best_epoch']
```

```python
epoch_idx, best_iou, best_epoch = load_checkpoints("checkpoint/heavyModel.pth", encoder, encoder_solver, decoder, decoder_solver,
                                                  refiner, refiner_solver, merger, merger_solver)
```

```python
def transform(image):
    
    assert (isinstance(image, np.ndarray))
    array = np.transpose(image, (0, 3, 1, 2))
    # handle numpy array
    tensor = torch.from_numpy(array)

    # put it from HWC to CHW format
    return tensor.float()
```

```python
def img2tensor(img):
    rendering_images = var_or_cuda(img)
    image = np.transpose(img, (2, 0, 1))
    img = np.transpose(img, (2, 0, 1))
    image = torch.from_numpy(img).unsqueeze(0)
    image = image.unsqueeze(0)
    return torch.tensor(image, dtype=torch.float32) 
```

```python
def generateVolume(image):
    with torch.no_grad():
        # Get data from data loader
        image_features = encoder(image)
        return decoder(image_features)
```

```python
def merge(raw_features, generated_volume):
    return merger(raw_features,generated_volume)
```

```python
def refine(generated_volume):
    with torch.no_grad():
        return refiner(generated_volume)
```

```python
#To the feture me when applying the features we just tested.....
#In the above function it does all the work for generating volumes of 1 view
#To add more than 1 view ... make sure that u merge the image with the features using merger
#After that use simple summation function to add merged volumes such the bellow function
#One last step is to refine the returned volume
#The last step is very important

def summation(volumes):
    gv = torch.zeros((1, 1, 32, 32, 32))
    for volume in volumes:
        gv = gv + volume
    return gv
```
